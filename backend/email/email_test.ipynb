```python
class Emailer2:
    def send_email(self, recipient, subject, content, image_paths=None):
        try:
            msg = MIMEMultipart()
            msg['From'] = GMAIL_USERNAME
            msg['To'] = recipient
            msg['Subject'] = subject

            # Create HTML content with images embedded
            html_content = f"<html><body>{content}</body></html>"
            msg.attach(MIMEText(html_content, 'html'))

            # Attach images
            if image_paths:
                for image_path in image_paths:
                    with open(image_path, 'rb') as image_file:
                        image_data = image_file.read()
                    image = MIMEImage(image_data)
                    image.add_header('Content-Disposition', 'attachment', filename=os.path.basename(image_path))
                    msg.attach(image)

            # Connect to Gmail Server
            session = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
            session.ehlo()
            session.starttls()
            session.ehlo()

            # Login to Gmail
            session.login(GMAIL_USERNAME, GMAIL_PASSWORD)

            # Send Email & Exit
            session.sendmail(GMAIL_USERNAME, recipient, msg.as_string())
            session.quit()
            print("Email sent")
        except Exception as e:
            print("Email could not be sent: ", e)
```

```python
class Emailer3:
    def send_email(self, recipient, subject, content, image_paths=None):
        try:
            msg = MIMEMultipart()
            msg['From'] = GMAIL_USERNAME
            msg['To'] = recipient
            msg['Subject'] = subject

            # Create HTML content with images embedded
            html_content = f"<html><body>{content}"
            
            if image_paths:
                for image_path in image_paths:
                    with open(image_path, 'rb') as image_file:
                        image_data = image_file.read()
                    image_base64 = base64.b64encode(image_data).decode('utf-8')
                    image_cid = os.path.basename(image_path)
                    html_content += f"<img src='data:image/jpeg;base64,{image_base64}' alt='{image_cid}'>"
            
            html_content += "</body></html>"
            msg.attach(MIMEText(html_content, 'html'))

            # Connect to Gmail Server
            session = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
            session.ehlo()
            session.starttls()
            session.ehlo()

            # Login to Gmail
            session.login(GMAIL_USERNAME, GMAIL_PASSWORD)

            # Send Email & Exit
            session.sendmail(GMAIL_USERNAME, recipient, msg.as_string())
            session.quit()
            print("Email sent")
        except Exception as e:
            print("Email could not be sent: ", e)
```

```python
Emailer3().send_email(fredi, emailSubject, emailContent, [bild])
```
