```python
import requests
```

```python
# Define the API endpoint
api_endpoint = 'http://api3.geo.admin.ch/rest/services/api/MapServer?searchText=wetter'

# Make the API request
try:
    response = requests.get(api_endpoint)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the response content (assuming it's in JSON format)
        data = response.json()

        # Now you can work with the 'data' variable containing the API response
        print(data)
    else:
        # Print an error message if the request was not successful
        print(f"Error: {response.status_code} - {response.text}")

except requests.exceptions.RequestException as e:
    # Handle exceptions such as network errors
    print(f"Request error: {e}")
```
